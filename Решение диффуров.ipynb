```python
def f(x,y,t):
    return y-1**2*sin(x)*t-0.05*y*t
def phase(n,t,x0,y0):
    xs=[]
    ys=[]
    x=x0
    y=y0
    for i in range(n):
        x1 = x+y*t
        y1 = f(x,y,t)
        x = x1
        y = y1
        xs.append(x)
        ys.append(y)

    plt.scatter(xs,ys)
```

```python
phase(5000,0.01,-5,3)
```
